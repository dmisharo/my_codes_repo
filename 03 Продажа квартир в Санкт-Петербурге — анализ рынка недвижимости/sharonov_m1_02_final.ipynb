```python
loc_without_distance = []
loc_dist_dic = {}
for name in tqdm(loc_name):
    a = data.loc[data['locality_name'] == name]['locality_name'].count()
    b = data.loc[(data['locality_name'] == name) & (data['city_centers_nearest'].isna())]['locality_name'].count()
    if a == b:
        loc_without_distance.append(name)
    else:    
        gr_median = data.loc[data['locality_name'] == name]['city_centers_nearest'].median()
        loc_dist_dic [name] = gr_median
```

С помощью библиотеки geopy будем определять координаты НП из списка 'loc_without_distance' и на их основе вычислять расстояние до 'spb_center_loc' 

```python
# создадим пустой массив для названий, которые не сможем определить
unknown_name = []
# создадим пустой словарь, куда будем сохранять пары навание-координаты
loc_coordinates = {}
# циклом перебираем названия НП
for name in tqdm(loc_without_distance):
        # в переменной new_name будем сохрянить новое наименование для уточнения поиска
        new_name = []
        for word in name.split():  # разбиваем имя на слова и проверяем первую букву слова
            if word[0] == word[0].upper(): 
                new_name.append(word) # если первая буква заглавная, добавляем слово в переменную
        new_name_reg = 'Ленинградская область, ' + (' '.join(new_name)) # создаём уточнённое название
        # возможны ошибки в определении координат. работаем через try
        try:
            location = app.geocode(new_name_reg).raw # получаем пакет геоданных по объекту
            distance = gd(spb_center_loc, (location["lat"], location["lon"])).m # расстояние до spb_center_loc в метрах
            loc_coordinates [name] = (location["lat"], location["lon"]) # добавляем в словарь имя и координаты
            loc_dist_dic [name] = distance # добавляем в словарь имя и расстояние
        # обработка ошибки. ели координаты найти не сможем, имя сохряняем в unknown_name и выводим сообщение об ошибки.  
        except:  
            unknown_name.append(name)
            print ('ERROR', name )    
```

Ошибок нет.

Добавляем значения из словаря  loc_dist_dic в данные перебирая названия НП с пропусками.

```python
for name in loc_name:
    mask = (data['locality_name'] == name) & data['city_centers_nearest'].isna()
    data.loc[mask, 'city_centers_nearest'] = loc_dist_dic.get(name) 
```

```python
# Меняем тип данных.
data['city_centers_nearest'] = data['city_centers_nearest'].astype('int')
# проверка
data.info()
```

Если для какого-либо объекта пропуски не удастся на что-то заменить, такие строки будем удалять. Строки с пропусками разобъем на 2 группы, объекты в 'Санкт-Петербург' и остальные. Так как по городу различия в удалённости от аэропорта , ввиду большой площади, могут быть существенны, замена пропусков на медианное или среднее значение по группе, может исказить информацию о стоимости объекта в зависимости от удалённости. Поэтому, объекты с пропусками в этой группе удалим. По второй группе примем, что населенные пункты области относительно небольшие по площади и пропуски можно заменить на медианное значение, если его удастся получить.

```python
# Посмотрим на число пропусков по Спб.
data.loc[data['airports_nearest'].isna() & (data['locality_name'] == 'Санкт-Петербург')]['locality_name'].count()
```

Всего 24 объекта. Удалим эти строки.

```python
idx = data.loc[data['airports_nearest'].isna() & (data['locality_name'] == 'Санкт-Петербург')].index
data = data.drop(idx)
# проверка
data.loc[data['airports_nearest'].isna() & (data['locality_name'] == 'Санкт-Петербург')]['locality_name'].count()
```

Проверим для каждого НП ко-во пропусков. Если число пропусков равно числу записей по НП, то заменить не можем и сохраним название НП в новый список. Если cможем определить медиану для груупы, то сохраним ее значение в словарь.

```python
loc_without_a_distance = []
loc_a_dist_dic = {}
for name in tqdm(loc_name):
    a = data.loc[data['locality_name'] == name]['locality_name'].count()
    b = data.loc[(data['locality_name'] == name) & (data['airports_nearest'].isna())]['locality_name'].count()
    if a == b:
        loc_without_a_distance.append(name)
    else:    
        gr_median = data.loc[data['locality_name'] == name]['airports_nearest'].median()
        loc_a_dist_dic [name] = gr_median
```

```python
# Заменяем пропуски полученными значениями.
for name in tqdm(loc_name):
    mask = (data['locality_name'] == name) & data['airports_nearest'].isna()
    data.loc[mask, 'airports_nearest'] = loc_a_dist_dic.get(name) 
```
